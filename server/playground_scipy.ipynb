```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate

# Points de donn√©es discrets
x = np.linspace(0, 10, 20)
y = np.sin(x)

# Cr√©er une fonction d'interpolation
f = interpolate.interp1d(x, y)

# G√©n√©rer des points pour interpolation
x_new = np.linspace(0, 10, 100)  # Points plus fins pour l'interpolation
y_new = f(x_new)

# Tracer les points d'origine et la courbe interpol√©e
plt.plot(x, y, 'o', label='Points discrets')  # Points originaux
plt.plot(x_new, y_new, '-', label='Interpolation lin√©aire')  # Courbe interpol√©e

# Mettre en √©vidence la valeur interpol√©e pour x = 5.5
plt.scatter(5.5, f(5.5), color='red', zorder=5,
            label=f'Interpolation √† x=5.5 : {f(5.5):.3f}')

plt.title('Interpolation lin√©aire des points')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import linalg

# D√©finir les matrices A et b
A = np.array([[3, 2], [1, 4]])
b = np.array([5, 6])

# R√©soudre Ax = b
x = linalg.solve(A, b)
print("Solution de x:", x)

# Fonction pour afficher une matrice sous forme de heatmap
def plot_matrix(matrix, title):
    plt.imshow(matrix, cmap='viridis', interpolation='none')
    plt.colorbar()
    plt.title(title)
    plt.show()


# Repr√©senter la matrice A, b et la solution x
plot_matrix(A, "Matrice A")
# Redimensionner b pour qu'il soit affich√© comme une colonne
plot_matrix(b.reshape(-1, 1), "Vecteur b")
# Afficher la solution x comme une colonne
plot_matrix(x.reshape(-1, 1), "Solution x")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# D√©finir la fonction quadratique


def f(x):
    return (x - 3) ** 2


# Trouver le minimum
result = minimize(f, 0)

# Tracer la fonction
x_values = np.linspace(-1, 7, 400)  # G√©n√©rer des valeurs de x de -1 √† 7
y_values = f(x_values)  # Calculer les valeurs correspondantes de f(x)

plt.plot(x_values, y_values, label='f(x) = (x-3)^2')
plt.scatter(result.x, f(result.x), color='red', zorder=5,
            label='Minimum')  # Marquer le minimum
plt.title('Fonction quadratique et son minimum')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
