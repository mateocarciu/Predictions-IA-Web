Entranes un modle de machine learning, il est important de suivre ses performances pendant l'entranement. Tu peux visualiser les courbes d'apprentissage pour comprendre comment le modle s'amliore au fil des itrations, ou observer les pertes (loss) et les prcisions pour viter le sur-apprentissage (overfitting).

```python
epochs = [1, 2, 3, 4, 5]
accuracy = [0.7, 0.8, 0.85, 0.87, 0.9]
loss = [0.5, 0.4, 0.35, 0.3, 0.25]

plt.plot(epochs, accuracy, label='Accuracy', color='blue')
plt.plot(epochs, loss, label='Loss', color='red')
plt.title('Courbe d\'apprentissage')
plt.xlabel('Epochs')
plt.ylabel('Valeur')
plt.legend()
plt.show()
```

En classification, la matrice de confusion est un outil prcieux pour visualiser o le modle fait des erreurs. Elle montre les vrais positifs, faux positifs, faux ngatifs, et vrais ngatifs.

y_test : Les classes relles.
y_pred : Les classes prdites par un modle.
confusion_matrix : Cette fonction calcule la matrice de confusion, qui compare les prdictions avec les classes relles.
sns.heatmap : Visualise la matrice de confusion sous forme de carte de chaleur avec les valeurs annotes.
La diagonale montre les prdictions correctes, tandis que les cases hors de la diagonale montrent les erreurs de prdiction. 

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Exemple imaginaire de vraies classes (y_test) et de classes prdites (y_pred)
y_test = np.array([0, 1, 1, 0, 1, 0, 1, 0, 1, 1])  # Classes relles
y_pred = np.array([0, 1, 0, 0, 1, 0, 1, 0, 0, 1])  # Classes prdites

# Calculer la matrice de confusion
cm = confusion_matrix(y_test, y_pred)

# Tracer la matrice de confusion
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Matrice de confusion')
plt.xlabel('Classe prdite')
plt.ylabel('Classe relle')
plt.show()
```
