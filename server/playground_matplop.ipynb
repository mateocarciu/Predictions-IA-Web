Entraînes un modèle de machine learning, il est important de suivre ses performances pendant l'entraînement. Tu peux visualiser les courbes d'apprentissage pour comprendre comment le modèle s'améliore au fil des itérations, ou observer les pertes (loss) et les précisions pour éviter le sur-apprentissage (overfitting).

```python
epochs = [1, 2, 3, 4, 5]
accuracy = [0.7, 0.8, 0.85, 0.87, 0.9]
loss = [0.5, 0.4, 0.35, 0.3, 0.25]

plt.plot(epochs, accuracy, label='Accuracy', color='blue')
plt.plot(epochs, loss, label='Loss', color='red')
plt.title('Courbe d\'apprentissage')
plt.xlabel('Epochs')
plt.ylabel('Valeur')
plt.legend()
plt.show()
```

En classification, la matrice de confusion est un outil précieux pour visualiser où le modèle fait des erreurs. Elle montre les vrais positifs, faux positifs, faux négatifs, et vrais négatifs.

y_test : Les classes réelles.
y_pred : Les classes prédites par un modèle.
confusion_matrix : Cette fonction calcule la matrice de confusion, qui compare les prédictions avec les classes réelles.
sns.heatmap : Visualise la matrice de confusion sous forme de carte de chaleur avec les valeurs annotées.
La diagonale montre les prédictions correctes, tandis que les cases hors de la diagonale montrent les erreurs de prédiction. ​

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Exemple imaginaire de vraies classes (y_test) et de classes prédites (y_pred)
y_test = np.array([0, 1, 1, 0, 1, 0, 1, 0, 1, 1])  # Classes réelles
y_pred = np.array([0, 1, 0, 0, 1, 0, 1, 0, 0, 1])  # Classes prédites

# Calculer la matrice de confusion
cm = confusion_matrix(y_test, y_pred)

# Tracer la matrice de confusion
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Matrice de confusion')
plt.xlabel('Classe prédite')
plt.ylabel('Classe réelle')
plt.show()
```
