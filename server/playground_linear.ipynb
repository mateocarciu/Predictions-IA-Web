```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Cration du DataFrame avec les donnes fournies

df = pd.read_csv('appartements.csv')

# Sparation des variables indpendantes (X) et de la variable dpendante (y)
X = df[['surface']]
y = df['price']

# Cration du modle de rgression linaire
model = LinearRegression()
model.fit(X, y)

# Prdictions sur les mmes donnes
y_pred = model.predict(X)

# Visualisation de la rgression linaire
plt.scatter(X, y, color='blue', label='Donnes relles')
plt.plot(X, y_pred, color='red', label='Rgression linaire')
plt.title('Rgression linaire entre la surface et le prix')
plt.xlabel('Surface (m)')
plt.ylabel('Prix ()')
plt.legend()
plt.grid(True)
plt.show()

# Affichage des coefficients de la rgression
coef = model.coef_
intercept = model.intercept_

# Affichage des coefficients de la rgression
print("Coefficient :", model.coef_)
print("Intercept :", model.intercept_)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Calculer le MSE, RMSE et R
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y, y_pred)

# Afficher les rsultats dans la console
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"R: {r2}")

# Visualisation des mtriques dans un graphique
metrics = ['MSE', 'RMSE', 'R']
values = [mse, rmse, r2]

plt.bar(metrics, values, color=['blue', 'orange', 'green'])
plt.title('Mtriques de performance du modle')
plt.ylabel('Valeurs')
plt.grid(True)
plt.show()
```

la rgression linaire peut tout  fait prendre en compte plusieurs variables explicatives (caractristiques), comme la surface, le nombre de chambres, et le nombre de fentres. C'est ce qu'on appelle une rgression linaire multiple, o plusieurs variables indpendantes sont utilises pour prdire une variable dpendante (ici, le prix d'un appartement).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

df = pd.read_csv('appartements.csv')

# Sparation des variables indpendantes (X) et de la variable dpendante (y)
# On utilise plusieurs variables explicatives
X = df[['surface', 'nbRooms', 'nbWindows']]
y = df['price']  # Variable cible

# Cration du modle de rgression linaire
model = LinearRegression()
model.fit(X, y)

# Prdictions sur les mmes donnes
y_pred = model.predict(X)

# Calculer le MSE, RMSE et R
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y, y_pred)

# Afficher les rsultats dans la console
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"R: {r2}")

# Affichage des coefficients de la rgression
print("Coefficients :", model.coef_)
print("Intercept :", model.intercept_)

# Visualisation des mtriques dans un graphique
metrics = ['MSE', 'RMSE', 'R']
values = [mse, rmse, r2]

plt.bar(metrics, values, color=['blue', 'orange', 'green'])
plt.title('Mtriques de performance du modle')
plt.ylabel('Valeurs')
plt.grid(True)
plt.show()
```
